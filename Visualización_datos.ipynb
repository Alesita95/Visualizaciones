```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = '/content/drive/MyDrive/BOOTCAMP_CODING_DOJO/PRACTICANDO_PLATAFORMA/Data.xlsx'
df = pd.read_excel(data)
df.head()
```

# Ordenando Datos

```python
df_1 = df.melt(id_vars=['Sport'], var_name='Año', value_name='Participantes')
df_1
```

```python
df2 = df_1.groupby(['Año','Sport'])[['Participantes']].sum().reset_index()
df2
```

```python
años = sorted(df2['Año'].unique())
```

```python
fig = px.scatter(df2, x='Año', y='Participantes',
                 size='Participantes',
                 color='Sport',
                 labels={'Sport': 'Deportes','Año':'Año' },
                 title='Deporte más popular en EEUU',)
fig.update_layout(showlegend=True, title_x=0.5)
fig.update_xaxes(tickvals=años)

fig.show()
```

Se presenta el siguiente gráfico donde nos indica cual es el deporte que popular en EEUU, siendo el football el que lidera como deporte mas popular entre los años encuestados
